Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the dataset to the dataframe
breast_cancer_data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
breast_cancer_data_frame.head()
```

```python
# adding the target column to the data frame
breast_cancer_data_frame['label'] = breast_cancer_dataset.target
```

```python
# Let's look at our dataframe agian
breast_cancer_data_frame.tail()
```

```python
# number of rows and columns
breast_cancer_data_frame.shape
```

```python
# getting some info
breast_cancer_data_frame.info()
```

```python
# checking missing values

breast_cancer_data_frame.isnull().sum()
```

```python
# some statistical measures about the data
breast_cancer_data_frame.describe()
```

```python
# checking the distribution of Target Variable
breast_cancer_data_frame['label'].value_counts()
```

1 -->Benign

0 -->Malignant

```python
breast_cancer_data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = breast_cancer_data_frame.drop(columns='label', axis=1)
Y = breast_cancer_data_frame['label']
```

```python
print(X)
print(Y)
```

Splitting the data into training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

Building the Neural Network

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid'),
])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs= 10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy for the model on test data


```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
